```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
def calculate_beta_of_stocks(stock_list):
    data = yf.download(stock_list,start = '2012-01-01', end = '2022-02-01', interval = '1mo')
    data = data['Adj Close']
    stock_data = data.dropna()
    #calculating log returns 
    stock_returns = np.log(stock_data).diff()
    stock_returns.dropna(inplace = True)
    
    
    #Downloading monthly returns of S&P500 and risk free rate
    
    #Add your own path of excel in this place
    data_sp500_rf = pd.read_excel("C:\\Users\\Dell\\Documents\\Sohail's Documents\\FTEC6303 Umit Gurun\\Assignment 4\\Assigment4_rf_mkt.xls")
    data_sp500_rf.set_index('date', inplace = True)
    mask = mask = (data_sp500_rf.index >= '2012-01-01') & (data_sp500_rf.index <= '2022-01-01')
    data_sp500_rf = data_sp500_rf.loc[mask]
    
    #This step calculates the sum of all monthly returns for s&p 500 and t bill for give time period.
    cum_returns_sp500 = data_sp500_rf.iloc[:,0].sum()
    cum_returns_rf = data_sp500_rf.iloc[:,1].sum()
    
    #calculating beta of each company
    beta_dict = {}
    #formula of beta = (expected return of stock - risk free rate)/ (expected return of market - risk free rate)
    for stock in stock_returns.columns:
        stock_cum_returns = stock_returns[stock].sum()
        beta = (stock_cum_returns - cum_returns_rf) / (cum_returns_sp500 - cum_returns_rf)
        beta_dict[stock] = beta
    
    #Converting dictionary into a dataframe
    df_beta = pd.DataFrame(beta_dict.items(), columns = ['Company', 'Beta'])
    df_beta.set_index('Company', inplace = True)
    

    return df_beta
```

```python
#Parameters
stock_list = ['BA','AAPL','HD', 'BAC', 'SBUX', 'RTX','XOM','BBY','HOG','IHG','MSFT','LUV','GOOGL']

calculate_beta_of_stocks(stock_list)

#We can conclude that technolgy stocks are highly correlated with the market and HOG is the only stock moving oppposite
#to the market in the given stocks.
```
